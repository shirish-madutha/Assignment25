```python
""" Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib. """

# ans
""" Matplotlib is a popular data visualization library in Python that provides a wide range of tools for 
creating high-quality plots, graphs, and figures. It is widely used in various fields, including data
analysis, scientific research, and machine learning.

Here are some key features and reasons why Matplotlib is used:

Flexible and Extensible: Matplotlib offers a wide range of plotting options and customization capabilities,
allowing users to create a diverse range of plots and visualizations.

Publication-Quality Output: Matplotlib provides precise control over the appearance and layout of plots, 
enabling the creation of publication-ready figures with high-quality output.

Integration with NumPy and Pandas: Matplotlib seamlessly integrates with other scientific libraries like
NumPy and Pandas, allowing easy visualization of data stored in arrays or DataFrames.

Support for Multiple Plot Types: Matplotlib supports a variety of plot types, making it suitable for
different types of data and analysis tasks.

Some commonly used plots that can be created using the Pyplot module of Matplotlib include:

Line Plot: A line plot displays data points connected by straight lines. It is useful for visualizing trends
and patterns over continuous variables.

Scatter Plot: A scatter plot displays individual data points as markers on a two-dimensional plane. 
It is useful for visualizing the relationship between two continuous variables.

Bar Plot: A bar plot represents categorical data with rectangular bars, where the length of each bar 
corresponds to the value of the category. It is useful for comparing categories or displaying counts or
frequencies.

Histogram: A histogram visualizes the distribution of a single variable by dividing it into bins and 
displaying the frequency or count of data points in each bin.

Pie Chart: A pie chart displays categorical data as sectors of a circle, where each sector represents a
category and its angle corresponds to the proportion of that category. """
```

```python
""" Q2: What is a scatter plot? Use the following code to generate data for x and y.
Using this generated data plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot. """

# ans
""" 
A scatter plot is a type of data visualization that displays the relationship between two numerical variables.
It represents individual data points as dots (or markers) on a two-dimensional coordinate system, with one
variable represented on the horizontal axis (x-axis) and the other variable represented on the vertical axis (y-axis).

Each dot on the scatter plot represents a single data point, and its position on the graph corresponds to 
the values of the variables it represents. The position of the dot along the x-axis is determined by the
value of one variable, while the position along the y-axis is determined by the value of the other variable. """

import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Title")
plt.scatter(x,y)
plt.show()
```

```python
""" Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data: 
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150]) """

# ans
""" The subplot() function in Matplotlib is used to create a grid of subplots within a single figure.
It allows you to plot multiple plots in a single figure, organizing them in rows and columns.

Each subplot within the grid can be addressed by its row and column indices. The subplot() function takes 
three arguments: the total number of rows, the total number of columns, and the index of the current 
subplot being created. """

import numpy as np
import matplotlib.pyplot as plt

font={"c":"green","size":10}
x = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2,2,1)
plt.plot(x, y1)
plt.title("Line 1", loc="right", fontdict=font)

y2 = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2,2,2)
plt.plot(x, y2)
plt.title("Line 2", loc="right", fontdict=font)

y3 = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2,2,3)
plt.plot(x, y3)
plt.title("Line 3", loc="right", fontdict=font)

y4 = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2,2,4)
plt.plot(x, y4)
plt.title("Line 4", loc="right", fontdict=font)

plt.show()
```

```python
""" Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal 
bar plot. """

# ans
""" A bar plot, also known as a bar chart or bar graph, is a type of data visualization that uses 
rectangular bars to represent categorical data. 
Bar plots are useful for visualizing categorical data because they provide a clear visual representation of
the differences in values between categories. They can effectively display the magnitude, frequency, or
count of categories, making it easy to identify patterns, trends, or outliers in the data.
"""

import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company, profit)
plt.xlabel("company")
plt.ylabel("profit")

plt.show()
```

```python
""" Q5: What is a box plot? Why is it used? Using the following data plot a box plot."""

# ans
""" A box plot, also known as a box-and-whisker plot, is a statistical visualization that provides a summary
of the distribution of a dataset. It displays the median, quartiles, and potential outliers of the data. 
The plot consists of a box that represents the interquartile range (IQR), a line or whisker that extends to
the minimum and maximum non-outlier values, and individual data points that may be considered outliers.

Box plots are used to:

Identify the central tendency: The line inside the box represents the median, which indicates the central
value of the dataset.

Show the spread of the data: The length of the box represents the IQR, which encompasses the middle 50% of
the data. The box indicates the range where the majority of the data lies.

Detect potential outliers: The whiskers extend to the minimum and maximum values that are within a 
reasonable range. Data points beyond the whiskers are considered potential outliers.

Compare distributions: Box plots are useful for comparing the distributions of multiple datasets side by 
side. They can provide insights into differences in central tendency, spread, and presence of outliers. """

import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]
labels = ['Box 1', 'Box 2']

plt.boxplot(data, labels=labels)
plt.xlabel('Box')
plt.ylabel('Values')
plt.title('Box Plot')
plt.show()
```
